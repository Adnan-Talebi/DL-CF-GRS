```python
import pandas as pd
import pathlib
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rcParams['figure.figsize'] = (20, 10)

def draw_hist(dataset, groupsize, subfixs=[''], filterby='user', subtitle=''):
    dfsix = []
    dfsval = []
    for subfix in subfixs:
        df = pd.read_csv(f'../data/grupos/{dataset}/groups-{groupsize}{subfix}.csv')
        ww=df.filter(regex=filterby)
        ww=ww.melt(value_vars=ww.columns)
        gg = ww.groupby('value').count()
        dfsix.append(gg['variable'].index)
        dfsval.append(gg['variable'])
    #return ww.hist(column='value', bins=6040)
    
    for ix, val in zip(dfsix, dfsval):
        plt.bar(ix, val, align='center', alpha=0.5)
    
    plt.ylabel('num')
    plt.title(filterby + ' distribution.'+ subtitle, fontsize = 40)

    plt.show()
```

# Todos vs Train

```python
draw_hist('ml1m',7, ['', '-train'], subtitle='All vs Train')
```
