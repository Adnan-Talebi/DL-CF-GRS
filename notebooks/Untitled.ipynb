```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Valores de los parámetros del modelo
w1 = 3
w2 = 3
bias = 1

# Definir una malla de puntos en el rango deseado
x1_values = np.linspace(-5, 5, 100)
x2_values = np.linspace(-5, 5, 100)
X1, X2 = np.meshgrid(x1_values, x2_values)

# Calcular los valores de z usando los parámetros del modelo
Z = w1 * X1**2 + w2 * X2**2 + bias

# Calcular los valores de la función sigmoid en la malla de puntos
sigmoid_values = sigmoid(Z)

# Crear el mapa de calor
plt.figure(figsize=(20,10))
plt.grid()
#plt.contourf(X1, X2, sigmoid_values, cmap='coolwarm')
plt.contourf(X1, X2, Z, cmap='coolwarm')
plt.colorbar(label='sigmoid(z)')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Mapa de calor de la función sigmoid')
plt.show()
```
